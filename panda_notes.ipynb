```python
# creating new metrics and dataframe 
M5 = np.array([["no","factor1",0],
               ["yes","factor2",2], 
               ["no","factor3",0], 
               ["no","factor4",2],
               ["yes","factor1", 1], 
               ["no","factor2",3],
               ["yes","factor5",3],
               ["no","factor1",0],
               ["yes","factor5",3],
               ["yes","factor1",2],
               ["yes","factor5",1],
               ["no","factor1",2],
               ["no","factor2",0],
               ["yes","factor1",1],
               ["no","factor3",2]
              ])
df2 = pd.DataFrame(M5, columns=["Request","Factor","Impact"])
df2
```

```python
# Adding column of random data to df2
# create random uniform array of values with compatible shape
cs = np.random.uniform(low=-1, 
                       high=1, 
                       size=(M5.shape[0])) # create a random array 
print(cs.shape)

# Assign array to dataframe
df2["kll_index"] = cs 
df2
```

```python
# conditionally fetching data
# condition is a NumPy array of booleans.
condition = df2['Request'] == 'no'
condition
```

```python
# We can then use this array to select only the rows that are True
df2[condition]
```

```python
# multiple condition
df2["Impact"] = df2["Impact"].astype(int)
condition1 = df2["Impact"] >= 2  
condition2 = df2["kll_index"] < 0 
condition3 = df2["Factor"] == "factor1"
df2_subset = df2[(condition1|condition2)&condition3]
df2_subset
```

```python
# apply: df['column'].apply(func)
# func can be any function which returns a single output.
# suppose we have this mapping 
mapping = {"factor1":"NA", "factor2":"MX","factor3":"LX","factor4":"TY","factor5":"PSOX"}

# apply the mapping to the Service_type column and create a new one  
df2["EDX"] = df2["Factor"].apply(lambda x: mapping.get(x,"Not Found")) 

# display result 
df2
```

```python
# loc: change value in column conditionally
condition = df2["Factor"] == "factor1"
df2.loc[condition, "Request"] = "no" 
df2
```

```python
# sorting
# inplace=True argument to change the original dataframe
df2.sort_values(by=["kll_index"], 
                ascending=True, 
                inplace=True) 
df2
```

```python
# Resetting the index
df2.reset_index(inplace=True)
df2 
```

```python
# deleting index column
df2.drop(columns=["index"], inplace=True)
df2
```

```python
# Aggregations
df_agg = df2.groupby("EDX").sum().reset_index()
df_agg
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10)) 

# Obtain x and y arrays
x = df2["Impact"] 
y = df2["kll_index"]

# plot the scatterplot  
plt.scatter(x,y, color="blue", marker="*") 

# add title and labels  
plt.title("Consumption index vs Impact") 
plt.xlabel("Impact") 
plt.ylabel("Consumption_index")
```
