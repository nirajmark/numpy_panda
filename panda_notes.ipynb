```python
import numpy as np 
import pandas as pd 

# initialize 2D array using numpy
M = np.random.randn(10,5)
M
```

```python
# access the columns attribute 
df.columns
```

```python
# creating new metrics and dataframe 
M5 = np.array([["no","factor1",0],
               ["yes","factor2",2], 
               ["no","factor3",0], 
               ["no","factor4",2],
               ["yes","factor1", 1], 
               ["no","factor2",3],
               ["yes","factor5",3],
               ["no","factor1",0],
               ["yes","factor5",3],
               ["yes","factor1",2],
               ["yes","factor5",1],
               ["no","factor1",2],
               ["no","factor2",0],
               ["yes","factor1",1],
               ["no","factor3",2]
              ])
df2 = pd.DataFrame(M5, columns=["Request","Factor","Impact"])
df2
```

```python
# Adding column of random data to df2
# create random uniform array of values with compatible shape
cs = np.random.uniform(low=-1, 
                       high=1, 
                       size=(M5.shape[0])) # create a random array 
print(cs.shape)

# Assign array to dataframe
df2["kll_index"] = cs 
df2
```

```python
# conditionally fetching data
# condition is a NumPy array of booleans.
condition = df2['Request'] == 'no'
condition
```

```python
# We can then use this array to select only the rows that are True
df2[condition]
```

```python
# multiple condition
df2["Impact"] = df2["Impact"].astype(int)
condition1 = df2["Impact"] >= 2  
condition2 = df2["kll_index"] < 0 
condition3 = df2["Factor"] == "factor1"
df2_subset = df2[(condition1|condition2)&condition3]
df2_subset
```
